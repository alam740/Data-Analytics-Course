One example of small data: Reliability of physics. We want to know how to characterize the defect generation of red boxes formation rate in a transistor. 

MOSFET is a transistor that powers our smartphones. On the blue circle, there are two squares which represent source and drain. In between, there is a small thin region with little red boxes which are called gate oxide. 

How it works: You apply a voltage on the solid black line on the top, often huge amount of voltage(ten thousand times higher than the high voltage power lines). Over time, the bonds of the red boxes are broken and more red boxes are produce.

With more red boxes, they will accidentally got aligned (this can be seen on the red circle) and the substrate will connect directly with the gate. Hence, the transistor will short and not work. 

This shows the importance of small data as the main outcome in this experiment is to avoid selling defective phone. They will test in the lab beforehand and apply a large bias.

On the graph, we wanted to see when the short circuit current happens which makes the current jumps up over function of time.

In order to obtain the maximum voltage that the phone will not fail, we will apply bigger voltage to find the maximum voltage.

We start with applying voltage at the green line and take the average time the objects in the sample fails. The higher the voltage, the shorter the time the transistor will short. 

On the right-hand graph, we plot the data point and use extrapolation to predict the safe operating voltage for phones within a specific time frame.

We have to know when to take the straight or the curve line. Avoid overfitting the data as the outcome will gives a wrong direction.

We will learn possible outliers where some of the data may fail at a higher or lower than the average time of the rest of the data points.

Small careful experiments are necessary to do significant extrapolation. This shows the importance of small data.

Big data vs. small data

Big data is seen as available data and easy to obtain through social medias, videos, Census data and insurance information. For example, Facebook time usage. These are not controlled experiments but rather observational experiments. This will not influence by any bias. The main thing is to ask questions that the data can answer. Big data often do not need high precision answers,for example, Netflix recommendation improvement increases from 35% to 40% accuracy translate to hundreds of millions of dollars. 

Small data is hypothesis driven and obtained from carefully and good experimental designs. This can carry bias.
Example: Galileo Saturn ring experiment.

Fundamental advance in big data is driven by memory technology where we can store data to analyze them. One long memory data saving is magnetic tapes.

```python
#Multiple Choice Code
from IPython.display import Image, display, clear_output
from ipywidgets import widgets

def create_multipleChoice_widget(description, options, correct_answer):
    if correct_answer not in options:
        options.append(correct_answer)
    
    correct_answer_index = options.index(correct_answer)
    
    radio_options = [(words, i) for i, words in enumerate(options)]
    alternativ = widgets.RadioButtons(
        options = radio_options,
        description = '',
        disabled = False
    )
    
    description_out = widgets.Output()
    with description_out:
        print(description)
        
    feedback_out = widgets.Output()

    def check_selection(b):
        a = int(alternativ.value)
        if a==correct_answer_index:
            s = '\x1b[6;30;42m' + "Correct." + '\x1b[0m' +"\n" #green color
        else:
            s = '\x1b[5;30;41m' + "False. " + '\x1b[0m' +"\n" #red color
        with feedback_out:
            clear_output()
            print(s)
        return
    
    check = widgets.Button(description="submit")
    check.on_click(check_selection)
    
    
    return widgets.VBox([description_out, alternativ, check, feedback_out])
```

```python
#Questions
Q1 = create_multipleChoice_widget('Galileo did an experiment involving pendulum to determine its period of oscillation. Was his experimental results small data or big data?',['Small data','Big data','All of the above','None of the above'],'Small data')
Q2 = create_multipleChoice_widget('Explain why incorrectly fitting a distribution may lead to incorrect decision and prediction?',['cat','dog','mouse'],'dog')
Q3 = create_multipleChoice_widget('What is the most important philosophical change in statistics over the last 300 years? ',['blue','white','red'],'white')
Q4 = create_multipleChoice_widget('Give some examples of big data vs. small data before personal computers became widely available? ',['k','d'],'k')
Q5 = create_multipleChoice_widget('Using google Trends, plot the popularity of your research topic over the last decade?',['h','d'],'d')
```

```python
#Display questions
display(Q1)
display(Q2)
display(Q3)
display(Q4)
display(Q5)
```
